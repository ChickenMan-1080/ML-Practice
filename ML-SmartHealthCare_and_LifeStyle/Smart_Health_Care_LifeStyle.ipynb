# Encoding

```python
df_encoded = pd.get_dummies(df,columns=['gender'],dtype=int) #ไม่ใช้ drop_first เพราะ randomforest ใช้ตีความ แต่ถ้าเป็น regression จะเสี่ยง Dummy Variable Trap
df_encoded
```

## Split the data

```python
from sklearn.model_selection import train_test_split

df_X = df_encoded.drop(labels=['heart_disease','diabetes','stroke','health_risk_score'],axis=1)
df_y = df_encoded[['heart_disease','diabetes','stroke']]

X_train,X_test,y_train,y_test = train_test_split(df_X,df_y, test_size=0.2,random_state=42)

df_X
```

## Train Model 

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(class_weight='balanced')

rfc.fit(X_train,y_train)

y_pred = rfc.predict(X_test)
```

```python
y_pred
```

# Evaluation of model

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

Accuracy = accuracy_score(y_test,y_pred)

print(classification_report(y_test,y_pred))
print(f'Accuracy score : {Accuracy:.2f}')
```

# Save Model

```python
import pickle

file_RFC = 'Health_care_RFC.pkl'

with open(file_RFC,'wb') as file:
    pickle.dump(rfc,file)
```
